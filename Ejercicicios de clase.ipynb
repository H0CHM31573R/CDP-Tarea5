```python
#Multiples simulaciones
sims = 100
results = {}
while sims <= 10000000:
    resultados_simulacion = []
    x_simulados = []

    for iteracion in range(1,sims +1,1):
        x_simulado = random.gauss(5,1) # sampleo/muestreo de la variable x que se distribuye de forma normal con media 5 y desviacion 2
        x_simulados.append(x_simulado)
        error = random.gauss(0,0.1) 

        # aplicar el modelo para obtener un valor simulado de y
        y = 0.1*x_simulado + 0.3  + error #aplicar el modelo a la variable "simulada"

        resultados_simulacion.append(y) #guardar los resultados en una lista para analisis posterior
        results[sims] = (x_simulados, resultados_simulacion)

    EY = sum(resultados_simulacion)/sims # E(Y) ,esperanza o valor esperado de una variable aleatoria
    ingreso_maximo = max(resultados_simulacion)
    ingreso_minimo = min(resultados_simulacion)

    print(f"Resultados generales de {sims} simulaciones")
    print("El valor esperado o esperanza de la variable Y es E(Y) =" ,EY)
    print("El valor minimo de y es ",ingreso_minimo)
    print("El valor maximo de y es ",ingreso_maximo)
    print("")
    sims *= 10
```

```python
%matplotlib inline

fig, axs = plt.subplots(len(results.keys()), 2)
fig.set_size_inches(23, 20)
fig.suptitle('Histogramas y grafico de dispercion')
for i, key in enumerate(results.keys()):
    axs[i, 0].hist(results[key][1])
    axs[i, 0].set_title(f"Histograma de {key} simulaciones")
    axs[i, 0].set(xlabel='y', ylabel='Frecuencia')
    
    axs[i, 1].scatter(results[key][0], results[key][1])
    axs[i, 1].set_title(f"Disperscion de {key} simulaciones")
    axs[i, 1].set(xlabel='x', ylabel='y')
```

**Ejercicio**: En ML usamos la entropía cruzada como una forma de medir que tan bueno es un modelo de variables discretas a través de comparar la distribución de probabilidad que el modelo produce o predice, vs la distribución de probabilidad real dada por los datos de entrenamiento.

Podemos ver el siguiente ejemplo que define la forma en que se calcula la entropía cruzada y nos muestra un caso específico:

<img src="https://image.slidesharecdn.com/publishintroductiontodeeplearninginpythonandmatlab1-160502102437/95/introduction-to-deep-learning-in-python-and-matlab-54-638.jpg?cb=1462185644">

En este caso interpretamos así: El problema consiste en un modelo o algoritmo de ML que debe producir un vector de 3 elementos indicando la probabilidad de que ciertos datos X pertenezcan a una de 3 categorías.

* El modelo de ML produce un vector que indica que estima un 70% de probabilidad de que se trate de la categoría 0, 20% de probabilidad de que se trate de la categoría 1 y 10% de que se trate de la categoría 2.
* Los datos reales nos dicen que se trataba de un caso donde con total certeza se sabe que se trata de la categoría 0
* La entropía cruzada(a calcular en el ejercicio) nos indica que tan buena es la estimación del modelo, una EC de 0 es un modelo perfecto(en este caso un modelo que predice 100% de prob para la clase 0)

**Nota** 
* Aun que para calcular la entropía usamos logaritmos en base 2, en ML para calcular la entropía cruzada se usa logaritmo natural ya que con este se cumple el proposito **estimar que tanto se alejan las predicciones del modelo de ML de los datos reales** y es comunmente mas rápido de calcular en la computadora.
* Ya que estamos trabajando con vectores que representan distribuciones de probabilidad , podemos toparnos con lo que se conoce como : **sparse vectors**(vectores donde la mayoría de elementos son 0), esto puede producir problemas ya que le logaritmo de 0 no esta definido, tu solución debe tomar en cuenta esto y evitar que devuelva "nan"

```python
def cross_entropy(Y,Y_hat):
    return -sum(Y*np.log(np.clip(Y_hat, 1e-15, 1-1e-15))) #Usando clipping y epsilon por defecto de scikit learn
    
    

y  = np.array([1.0,0,0])
y_hat = np.array([0.7,0.2,0.1]) # <- resultado de algun modelo predictivo

cross_entropy(y,y_hat)
```

**Ejercicio aplicado en DS**: la correlación cruzada es una medida de similitud entre 2 funciones como resultado de "desplazar" una sobre la otra, comunmente es usada para encontrar características relevantes en una función desconocida o no controlada.

En procesamiento de señales por ejemplo es usada para buscar y/o filtrar en una señal que varia en el tiempo cierta caracaterística de interés. Aveces es llamada también "sliding dot product" consiste en aplicar en cada punto de una función F el producto punto con cierta función G(comunmente mas corta) y luego "deslizar" G a un nuevo punto de la función, el resultado es una nueva función H que se interpreta como :**cuanto se parece** en cada punto la función F a la característica G. 

<img src="https://i.makeagif.com/media/11-25-2015/LZ9Ufj.gif">

En este ejercicio usamos correlación cruzada para calcular las medias moviles promediando 3 puntos que ya vimos en otro ejemplo:

El primer paso es definir la función G que define el "filtro" a aplicar, para este caso consiste simplemente en un vector con 3 elementos donde cada elemento corresponde a 1/3, luego debemos aplicar el producto punto sobre cada punto de la función o datos originales(en este ejemplo llamados x) a traves aplicar el producto punto en cada elemento "corriendolo" de uno en uno

```python
ruido = 0.1*np.random.randn(15) #el ruido comunmente se debe a aleatoriedad o captura no exacta de info.
x = np.linspace(0,2*np.pi,15) 
y = np.sin(x) + ruido

filtro = np.array([1/3,1/3,1/3]) #para lograr las medias moviles a traves de correlacion cruzada

def naive_sliding_dot_product(y, fil):
    test = [y[0]]
    for i in range(0, len(y) - len(fil) + 1):
        test.append(np.dot(y[i:i+len(fil)],fil))
    test = test + [y[len(y)-1]]
    return np.array(test)
    

plt.plot(x,y)
plt.plot(x,naive_sliding_dot_product(y, filtro))
plt.show()
```

**ejercicio** Implementar la función:
$$h(x) = \begin{cases}0 & x< 0\\1 & 0<=x <=1\\0 & x> 1\end{cases}$$

Esta debe funcionar para vectores de cualquier tamaño x:

def h(x):

    ...

**nota** debe ser implementada sin ciclos o ifs 

```python
def h(x):
    h = np.copy(x)
    h[np.logical_and(h>=0, h<=1)] = 1
    h[np.logical_or(h<0, h>1)] = 0
    return h

x = np.array([0.1,-2,0.5,5])
print(h(x))
```

**Ejercicio**

Usando select implementar una función k(x) con la siguiente definición:
$$k(x) = \begin{cases}-x & x< 0\\x^{3}  &0<=x<1\\x^{2}  &1<=x<2\\4  &otherwise\end{cases}$$



```python
def k(x):
    k = np.copy(x)
    conditionals = [k<0, np.logical_and(k>=0, k<1), np.logical_and(x>=1, x<2)]
    results =      [-k , k**3                     , k**2]
    return np.select(conditionals, results, 4)

k(x)
```

##### Select y performance
Aun que select es muy conveniente y útil , no es la opción mas eficiente esto debido a que evalua todas las condiciones y todos los resultados, en programación en general buscamos escribir los programas de la manera mas eficiente posible y evitar calculos innecesarios, esto se vuelve vital en ciencia de datos cuando procesamos grandes volúmenes de información, por eso NumPy nos provee la alternativa a select llamada **piecewise** que funciona de manera similar .


Este funciona similar a select pero en vez de calcular todos los posibles resultados, calcula solo aquellos para los que la condición es True e ignora los False,sintácticamente piecewise requiere que los "resultados" sesan calcuados usando una lista de funciones por lo cual todos los resultados deben estar contenidos en una función  y el objeto función ser enviado a piecewise (si la función solo se utiliza una vez para este propósito, se puede usar funciones anónimas o lambda).

La diferencia principal, y de la cual se derivan el resto de diferencias, es que los arrays de numpy son de n-dimensiones (de hecho, realmente se llaman ndarray), mientras que las matrices son estrictamente de dos dimensiones. Las np.matrix son estrictamente una subclase (caso especifico) de los arrays (caso general), por lo tanto se benefician de tener comportamientos particulares y especificos para su caso de uso.

Utilizando multiplicación de matrices se debe calcular para cada fila de la matriz de entrada X el valor de las neuronas de la capa intermedia, esto producirá una nueva matriz con el mismo número de filas que X y 2 columnas(1 para cada neurona) , a  los valores de esta matriz se les debe aplicar la función "sigmoid"(descrita a continuación) para limitarlos al intervalo de 0 a 1, esto produce una matriz del mismo tamaño pero con valores entre 0 a 1, esta matriz se multiplica matricialmente por la matriz que representa los pesos de la capa de salida  y este proceso produce un nuevo tensor al cual se debe aplicar nuevamente la función sigmoid. El resultado debe ser un tensor con el mismo número de filas que la matriz X y una sola columna(una predicción para cada fila de X.

```python
# Aplicar la red neuronal sobre los siguientes datos X

X1 = np.array([0.50,0.72])
X2 = np.array([-4,7])
X3 = np.zeros_like(X2)
x4 = np.ones_like(X1)
X5 = np.random.randn(X1.shape[0])
exesmat = np.matrix([X1,X2,X3,x4,X5])

def sigmoid(x): #convertir los valores de x al rango de 0 a 1    
    return 1/(1+np.exp(-x))

def neural_network(x):
    hlmat = np.mat([[0.25, 0.37], [-8, 14]])
    output = np.mat([4, 9])
    sigproduct = sigmoid(x*hlmat.T)
    return sigmoid(sigproduct*output.T)

neural_network(exesmat)
```
